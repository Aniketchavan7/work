```python
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df['normalized-losses'].fillna(df['normalized-losses'].mean())
```

```python
colums_replace=['bore','stroke','horsepower','peak-rpm']
for i in colums_replace:
    df[i]=pd.to_numeric(df[i],errors='coerce')
    df[i].replace(np.NaN, df[i].mean())
```

```python
df.dropna(subset=['price'],axis=0)
df['price']=pd.to_numeric(df['price'],errors='coerce')
df['price'].fillna(df['price'].mean())
df.isnull().sum()
```

```python
df['city-mpg']=pd.to_numeric(df['city-mpg'],errors='coerce')
df['highway-mpg']=pd.to_numeric(df['highway-mpg'],errors='coerce')
```

```python
df['city-mpg']=235/df['city-mpg'].astype('float')
df['highway-mpg']=235/df['highway-mpg'].astype('float')
df.rename(columns={'highway-mpg':'highway-kmph','city-mpg':'city-kmph'})
```

```python
cols=['length','width','height']
for i in cols:
    df[i]=pd.to_numeric(df[i],errors='coerce')
```

```python
df['length']=df['length']/df['length'].max()
df['height']=df['height']/df['height'].max()
df['width']=df['width']/df['width'].max()
```

```python
df['num-of-doors']=df['num-of-doors'].replace({'four':'4','two':'2'})
```

```python
abc={'std':0,'turbo':1}
df['aspiration']=df['aspiration'].map(abc)
```
